## Wikipedia Metin ve Görsel İşleme

### Dosya Okuma

```python
import pandas as pd
```

```python
df = pd.read_csv("wiki_data.csv")
df.head()
```

```python
def clean_text(df):
    df["text"] = df["text"].str.lower()
    df["text"] = df["text"].str.replace('[^\w\s]', '', regex=True)
    df["text"] = df["text"].str.replace('\d', '', regex=True)
```

```python
clean_text(df)
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
sw = stopwords.words('english')
```

```python
def remove_stopwords(df, stopwords):
    df["text"] = df["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords]))
    
```

```python
remove_stopwords(df, sw)
```

```python
df.head()
```

```python
temp_df = pd.Series(' '.join(df['text']).split()).value_counts()
temp_df
```

```python
drops = temp_df[temp_df <= 2000]
drops
```

```python
df["text"] = df["text"].apply(lambda x: ' '.join([word for word in str(x).split() if word not in (drops)]))
df.head()
```

```python
from textblob import Word

nltk.download('wordnet')
```

```python
df['text'] = df['text'].apply(lambda x: ' '.join([Word(word).lemmatize() for word in x.split()]))
df.head()
```
